```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame
from sklearn import datasets
from sklearn.mixture import GaussianMixture

# load the iris dataset
iris = datasets.load_iris()

# select first two columns
X = iris.data[:, :2]

# turn it into a dataframe
d = pd.DataFrame(X)

# plot the data
plt.scatter(d[0], d[1])
plt.show()
```

```python
gmm = GaussianMixture(n_components = 3)

# Fit the GMM model for the dataset
# which expresses the dataset as a
# mixture of 3 Gaussian Distribution
gmm.fit(d)

# Assign a label to each sample
labels = gmm.predict(d)
d['labels']= labels
d0 = d[d['labels']== 0]
d1 = d[d['labels']== 1]
d2 = d[d['labels']== 2]

# plot three clusters in same plot
plt.scatter(d0[0], d0[1], c ='r')
plt.scatter(d1[0], d1[1], c ='yellow')
plt.scatter(d2[0], d2[1], c ='g')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans

data1 = pd.read_csv('/content/cluster_data.csv')
data2 = pd.read_csv('/content/cluster_data1.csv')

features_data1 = data1.values
features_data2 = data2.values

n_components = 2
gmm_data1 = GaussianMixture(n_components=n_components, random_state=42)
labels_data1 = gmm_data1.fit_predict(features_data1)

kmeans_data1 = KMeans(n_clusters=2, random_state=42)
kmeans_labels_data1 = kmeans_data1.fit_predict(features_data1)

gmm_data2 = GaussianMixture(n_components=n_components, random_state=42)
labels_data2 = gmm_data2.fit_predict(features_data2)

kmeans_data2 = KMeans(n_clusters=2, random_state=42)
kmeans_labels_data2 = kmeans_data2.fit_predict(features_data2)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(features_data1[:, 0], features_data1[:, 1], c=labels_data1, cmap='viridis', s=40, alpha=0.6)
plt.scatter(gmm_data1.means_[:, 0], gmm_data1.means_[:, 1], marker='x', s=100, color='red', label='GMM Centroids')
plt.title('Gaussian Mixture Model for data1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(features_data1[:, 0], features_data1[:, 1], c=kmeans_labels_data1, cmap='viridis', s=40, alpha=0.6)
plt.scatter(kmeans_data1.cluster_centers_[:, 0], kmeans_data1.cluster_centers_[:, 1], marker='x', s=100, color='red', label='K-means Centroids')
plt.title('K-means Clustering for data1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

plt.show()

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(features_data2[:, 0], features_data2[:, 1], c=labels_data2, cmap='viridis', s=40, alpha=0.6)
plt.scatter(gmm_data2.means_[:, 0], gmm_data2.means_[:, 1], marker='x', s=100, color='red', label='GMM Centroids')
plt.title('Gaussian Mixture Model for data2')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(features_data2[:, 0], features_data2[:, 1], c=kmeans_labels_data2, cmap='viridis', s=40, alpha=0.6)
plt.scatter(kmeans_data2.cluster_centers_[:, 0], kmeans_data2.cluster_centers_[:, 1], marker='x', s=100, color='red', label='K-means Centroids')
plt.title('K-means Clustering for data2')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

plt.show()



```

```python
# Import scipy and draw the
# line of Linear Regression:


import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Change 'your_data.csv' to the path of your CSV file
your_data = pd.read_csv('/content/cluster_data1.csv')

# Display the first few rows of the loaded data
print(your_data.head())

# Adjust column indices based on your data
X = your_data.iloc[:, :]


plt.scatter(X.iloc[:, 0], X.iloc[:, 1])
plt.title('Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

gmm = GaussianMixture(n_components=3)
gmm.fit(X)


labels = gmm.predict(X)
your_data['labels'] = labels

# Separate data based on labels
cluster0 = your_data[your_data['labels'] == 0]
cluster1 = your_data[your_data['labels'] == 1]
cluster2 = your_data[your_data['labels'] == 2]

# Plot clusters
plt.scatter(cluster0.iloc[:, 0], cluster0.iloc[:, 1], c='r', label='Cluster 0')
plt.scatter(cluster1.iloc[:, 0], cluster1.iloc[:, 1], c='yellow', label='Cluster 1')
plt.scatter(cluster2.iloc[:, 0], cluster2.iloc[:, 1], c='g', label='Cluster 2')

plt.title('Gaussian Mixture Model Clusters')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
